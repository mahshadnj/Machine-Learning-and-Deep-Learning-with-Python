**Problem 1:**
Plot a sin function as well as noise corrupted data from this sin function.

```python
import numpy as np
import matplotlib.pyplot as plt

# creating the sine function
x = np.linspace(0, 1, 100)
y = np.sin(2 * np.pi * x)

# Generate noise corrupt data from the sine function
x_noisy = np.linspace(0, 1, 10)
y_noisy = np.sin(2 * np.pi * x_noisy) + np.random.normal(0, 0.1, x_noisy.shape)

plt.plot(x, y, label='sine function')
plt.scatter(x_noisy, y_noisy, color='blue', label='observed noise data')
plt.title("Plot of noise corrupted data")

plt.show()
```

**Problem 2:**
Determine if an array is monotonic. (Without using Python function)

Example:

- [1,2,3,3,3,3,3,4,5,6,7] is monotonic (increasing)
- [7,6,5,4,3,3,3,3,3,2,1] is monotonic (decreasing)
- [1,5,4,3,3,3,3,3,2,6,7] is not monotonic

```python
def is_monotonic(array):
    increasing = True
    decreasing = True

    for i in range(1, len(array)):
        if array[i] < array[i - 1]:
            increasing = False
        if array[i] > array[i - 1]:
            decreasing = False

    if increasing:
        print(f"{array} is monotonic (increasing)")
    elif decreasing:
        print(f"{array} is monotonic (decreasing)")
    else:
        print(f"{array} is not monotonic")

# Example test cases
is_monotonic([1, 2, 3, 3, 3, 3, 4, 5, 6, 7])      # increasing
is_monotonic([7, 6, 5, 4, 3, 3, 3, 3, 2, 1])      # decreasing
is_monotonic([1, 5, 4, 3, 3, 3, 3, 2, 6, 7])      # not monotonic
```
